```python
import ipywidgets as widgets
from IPython.display import display

# Portas L√≥gicas
def AND(a, b):
    return a & b

def OR(a, b):
    return a | b

def NOT(a):
    return 1 - a

# Circuito
def circuito(A, B, C):
    parte1 = AND(A, B)
    parte2 = NOT(C)
    S = OR(parte1, parte2)
    return parte1, parte2, S

# Widgets (Bot√µes)
A = widgets.ToggleButtons(options=[0,1], description='A:')
B = widgets.ToggleButtons(options=[0,1], description='B:')
C = widgets.ToggleButtons(options=[0,1], description='C:')

output = widgets.Output()

def atualizar(change):
    output.clear_output()
    parte1, parte2, S = circuito(A.value, B.value, C.value)

    with output:
        print("Parte 1 (A AND B) =", parte1)
        print("Parte 2 (NOT C) =", parte2)
        print("Sa√≠da Final S =", S)

A.observe(atualizar, names='value')
B.observe(atualizar, names='value')
C.observe(atualizar, names='value')

display(A, B, C, output)
```

```python
import random
import ipywidgets as widgets
from IPython.display import display, clear_output

# Portas L√≥gicas
def AND(a,b): return a & b
def OR(a,b): return a | b
def XOR(a,b): return a ^ b
def NAND(a,b): return 1 - (a & b)
def NOR(a,b): return 1 - (a | b)

portas = {
    "AND": AND,
    "OR": OR,
    "XOR": XOR,
    "NAND": NAND,
    "NOR": NOR
}

# Escolhe porta secreta
porta_secreta_nome = random.choice(list(portas.keys()))
porta_secreta = portas[porta_secreta_nome]

# Bot√µes
A = widgets.ToggleButtons(options=[0,1], description='A:')
B = widgets.ToggleButtons(options=[0,1], description='B:')
saida = widgets.HTML(value="<h3>Sa√≠da: ?</h3>")

palpite = widgets.Dropdown(
    options=list(portas.keys()),
    description='Qual √© a porta?'
)

botao_verificar = widgets.Button(description="Verificar")

resultado = widgets.Output()

# Atualiza sa√≠da
def atualizar(change):
    s = porta_secreta(A.value, B.value)
    if s == 1:
        saida.value = "<h3 style='color:red;'>Sa√≠da: 1 üî¥</h3>"
    else:
        saida.value = "<h3 style='color:green;'>Sa√≠da: 0 üü¢</h3>"

A.observe(atualizar, names='value')
B.observe(atualizar, names='value')

# Verifica resposta
def verificar(b):
    resultado.clear_output()
    with resultado:
        if palpite.value == porta_secreta_nome:
            print("üéâ ACERTOU! Era", porta_secreta_nome)
        else:
            print("‚ùå Errado! Era", porta_secreta_nome)

botao_verificar.on_click(verificar)

display(A, B, saida, palpite, botao_verificar, resultado)
```
